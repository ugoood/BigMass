## 2. Basic usage
### 2.1 Draw a quadratic curve

````python
# %matplotlib inline        # 省去了以后的 plt.show(), 用在jupyter notebook中
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(-1,1,50)     # 生产-1 to 1 之间的 50 points
# y = 2*x +1   # line
y = x**2          # quadratic curve
````

````python
plt.plot(x, y)
````

````python
# plt.show()      # 脚本中 得用这句话，否则不出图
````

## 2.2 Figure image
We prepare to draw two figures seperately:

````python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure()
plt.plot(x, y1)

# 注意draw 两个 figure 的方式
# 默认 figure 1,  figure 2, 但是可以用 num = 3 定义成 figure3
# pyhon2 报错 plt.figure(num = 3, figsize(8, 5))  # figsize param 定义图width & length
plt.figure(num = 3)
plt.plot(x, y2)

plt.show()
````

两种线画在同一个图里：

````python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure(num = 4)
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 5.0, linestyle = '--')

plt.show()
````

## 2.3 Figure axis 1 设置坐标轴1

````python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure(num = 4)
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')

plt.xlim((-1, 2))      # 设置 水平坐标轴
plt.ylim((-2, 3))
plt.xlabel("I am x")      # 设定 axis 标签
plt.ylabel("I am y")

new_ticks = np.linspace(-1, 2, 5)      
print(new_ticks)
plt.xticks(new_ticks)        # 换 x axis 的 ticks

plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$really\ bad$', r'$bad\ \beta$',   r'normal\ \alpha', 'good', r'$really\ good$'])
# 上面， “r” 表示正则， “\” 为转义符，想把空格显示出来。
# 没$符号，打不出alpha

plt.show()
````
