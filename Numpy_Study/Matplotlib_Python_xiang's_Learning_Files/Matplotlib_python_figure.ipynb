### 2.2 Figure image
We prepare to draw two figures seperately:

````python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure()
plt.plot(x, y1)

# 注意draw 两个 figure 的方式
# 默认 figure 1,  figure 2, 但是可以用 num = 3 定义成 figure3
# pyhon2 报错 plt.figure(num = 3, figsize(8, 5))  # figsize param 定义图width & length
plt.figure(num = 3)
plt.plot(x, y2)

plt.show()
````

### 2.3 Figure axis 1 设置坐标轴1

````python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure(num = 4)
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')

plt.xlim((-1, 2))      # 设置 水平坐标轴
plt.ylim((-2, 3))
plt.xlabel("I am x")      # 设定 axis 标签
plt.ylabel("I am y")

new_ticks = np.linspace(-1, 2, 5)      
print(new_ticks)
plt.xticks(new_ticks)        # 换 x axis 的 ticks

plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$really\ bad$', r'$bad\ \beta$',   r'normal\ \alpha', 'good', r'$really\ good$'])
# 上面， “r” 表示正则， “\” 为转义符，想把空格显示出来。
# 没$符号，打不出alpha

plt.show()
````

### 2.4 Figure axis 2 设置坐标轴2
修改坐标轴的位置：

````python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure(num = 4)
plt.plot(x, y2)
plt.plot(x, y1, color = 'red', linewidth = 1.0, linestyle = '--')

plt.xlim((-1, 2))      # 设置 水平坐标轴
plt.ylim((-2, 3))
plt.xlabel("I am x")      # 设定 axis 标签
plt.ylabel("I am y")

new_ticks = np.linspace(-1, 2, 5)      
print(new_ticks)
plt.xticks(new_ticks)        # 换 x axis 的 ticks

plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$really\ bad$', r'$bad\ \beta$',   r'$normal\ \alpha$', 'good', r'$really\ good$'])
# 上面， “r” 表示正则， “\” 为转义符，想把空格显示出来。
# 没$符号，打不出alpha

# gca = 'get current axis'
ax = plt.gca()    
ax.spines['right'].set_color('none')    # 轴， 图片的四个边框
ax.spines['top'].set_color('none')

# 设置x 和 y 的默认坐标轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 移动x，y轴的位置
ax.spines['bottom'].set_position(('data', 0))   # x轴上移到 y轴 = 0的位置
ax.spines['left'].set_position(('axes', 0.1))  # y轴移动到x轴数据10%的位置

# set_position里参数有[ 'outward' | 'axes' | 'data' ]
# data: 表那个轴数据的位置
# axes: 表哪个轴数据的百分比位置
# outward:

plt.show()
````

### 2.5 Legend 图例

````python
x = np.linspace(-3,3,50)
y1 = 2*x +1
y2 = x**2

plt.figure(num = 4)
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel("I am x")
plt.ylabel("I am y")

new_ticks = np.linspace(-1, 2, 5)      
print(new_ticks)
plt.xticks(new_ticks)

plt.yticks([-2, -1.8, -1, 1.22, 3], 
           [r'$really\ bad$', r'$bad\ \beta$',  
            r'normal\ \alpha', 'good', r'$really\ good$'])

l1, = plt.plot(x, y2, label = 'up')    # 给线取名字
l2, = plt.plot(x, y1, color = 'red', linewidth = 1.0, 
         linestyle = '--', label = 'down')

# 打印出图例
plt.legend(handles = [l1,l2,], labels =['aaa', 'bbb'] , loc = 'best')   
# loc = 'best'  总能把legend调整到适当的地方
# loc = 'upper right'      # 'lower right' 等多种形式

#试一试：
# plt.legend(labels =['aaa', 'bbb'] , loc = 'best')   
# plt.legend(labels =['aaa', 'bbb'] )   
# plt.legend(labels =['aaa'] )   

plt.show()
````

### 2.6  Annotation
添加两种形式的annotation.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y = 2*x + 1

plt.figure(num = 1)     # 报错plt.figure(num = 1, figsize(8,5),)
plt.plot(x,y)       # 试一试plt.scatter

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

x0 = 1
y0 = 2*x0 + 1

plt.scatter(x0, y0, s = 40, color = 'r')
plt.plot([x0, x0], [y0, 0], 'k--', lw = 2.5)    # 两点(x0, y0), (x0, 0)之间画线

# method 1
##############
plt.annotate(r'$2x+1=%s$'%y0, xy=(x0, y0), 
             xycoords = 'data', xytext = (+30, -30), 
             textcoords = 'offset points', fontsize = 16,
            arrowprops = dict(arrowstyle = '->', 
                              connectionstyle = 'arc3, rad = .2'))

# method 2
##############
plt.text(-5.5, y0, r'$This\ is\ the\ some\ text,\ \mu\ \sigma_i\ \alpha_t$',
        fontdict = {'size':16, 'color':'r'})

plt.show()
````

### 2.7 tick 能见度
如果图片复杂，挡住了坐标轴的ticks, 可选择本节内容。

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y = 0.1*x

plt.figure(num = 1)
plt.plot(x,y, linewidth = 10)
plt.ylim(-2, 2)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor = 'white', edgecolor = 'None',
                       alpha = 0.7))

plt.show()
````

## 3. Figure types
### 3.1 Scatter

````python
import matplotlib.pyplot as plt
import numpy as np

n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)   # for color value

# cmap参数是把x,y映射到颜色里； alpha是透明度, s = size
plt.scatter(X, Y, s =75, c = T, alpha = 0.5)     
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
plt.xticks(())         # 去掉x上的ticks
plt.yticks(())

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# scatter's general usagge
plt.scatter(np.arange(5), np.arange(5))
plt.show()
````

### 3.2 Bar

````python
import matplotlib.pyplot as plt
import numpy as np

n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.figure(num = 1)
##### bar #####
plt.bar(X, +Y1, facecolor = '#9999ff', edgecolor = 'w')
plt.bar(X, -Y2, facecolor = '#ff9999', edgecolor = 'white')

for x, y in zip(X, Y1):
    # ha: horizontal alignment
    # va
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha = 'center', va = 'bottom')
    
for x, y in zip(X, Y2):
    # ha: horizontal alignment
    # va
    plt.text(x+0.4, -y-.05, '%.2f' % -y, ha = 'center', va = 'top')


#############


plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())

plt.show()
````

### 3.3 Contours 等高线图

````python
import matplotlib.pyplot as plt
import numpy as np

def f(x, y):
    # the heiht function
    return (1 - x / 2 + x**5 + y**3) * np.exp(-x**2 - y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)

X, Y = np.meshgrid(x, y)

m = 8
# use plt.contourf to filing contours
plt.contourf(X, Y, f(X, Y), m, alpha = 0.75, cmap = plt.cm.hot)

# use plt.contours to add contour lines
C = plt.contour(X, Y, f(X, Y), m, colors = 'black', linewidth = 0.5)

# add label
plt.clabel(C, inline = True, fontsize = 8)

plt.xticks(())
plt.yticks(())
plt.show()
````

### 3.5 3-D plot

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

# setting X, Y
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.15)
X, Y = np.meshgrid(X, Y)

R = np.sqrt(X**2 + Y**2)

# height value
Z = np.sin(R)

# rstride, cstride：     row, column 跨
ax.plot_surface(X,Y,Z, rstride = 1, 
                cstride = 1, cmap = plt.get_cmap('rainbow'))

# zdir = 'z' 表示顺着z轴的投影
ax.contourf(X, Y, Z, zdir = 'z', offset = -2, cmap = 'rainbow')

ax.set_zlim(-2, 2)
plt.show()
````

### 4.3 Plot in plot

````python
import matplotlib.pyplot as plt

fig = plt.figure()
x = [1, 2, 3, 4, 5, 6, 7]
y = [1, 3, 4, 2, 5, 8, 6]

left, bottom, width, height = 0.1, 0.1, 0.8, 0.8
ax1 = fig.add_axes([left, bottom, width, height])
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

left, bottom, width, height = 0.2, 0.6, 0.25, 0.25
ax2 = fig.add_axes([left, bottom, width, height])
ax2.plot(x, y, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title inside 1')

plt.axes([.6, .2, .25, 0.25])
plt.plot(x, y[::-1], 'g')
plt.xlabel('x')
plt.ylabel('y')
plt.title("title inside 2")

plt.show()
````

### 4.4 Secondary axis

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y1 = 0.05 * x**2
y2 = -1*y1

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()      # 镜面
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2,'r--')

ax1.set_xlabel('X data')
ax1.set_ylabel(r'$I\ am\ Y_1$', color = 'g')
ax2.set_ylabel('Y2', color = 'r')


plt.show()
````

## 5 Animation

````python
# jupyter notebook 中显示不出来， 可在terminal中执行
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x+i/10))
    return line,

def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig = fig, func = animate, 
                              frames = 100, init_func = init, interval = 20, blit = False)
plt.show()
````
